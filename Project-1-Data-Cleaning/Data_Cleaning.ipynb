```python
df['CouponCode']=df['CouponCode'].fillna('No Coupon')
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce').dt.strftime('%Y-%m-%d')
```

```python
df.dtypes
```

```python
cols = ['Product', 'PaymentMethod', 'OrderStatus', 'ReferralSource',
             'ShippingAddress', 'CustomerID', 'TrackingNumber', 'CouponCode']

for col in cols:
    df[col] = df[col].str.strip()   # removes leading/trailing spaces only

```

```python
df['UnitPrice'] = df['UnitPrice'].round(2)
df['TotalPrice'] = df['TotalPrice'].round(2)
```

```python
(df['TotalPrice'].round(2) != (df['Quantity']*df['UnitPrice']).round(2)).sum()
```

```python
total = len(df)
duplicates = df['OrderID'].duplicated().sum()
error_rate = (duplicates / total) * 100

print(f"Duplicate OrderIDs: {duplicates} out of {total}")
print(f"Error rate: {error_rate:}%")
```

```python
data = pd.to_datetime(df['Date'], format='%Y-%m-%d', errors='coerce')
wrong_dates = data.isnull().sum()
error_rate = (wrong_dates / total) * 100

print(f"Invalid formatted dates: {wrong_dates} out of {total}")
print(f"Error rate: {error_rate:}%")
```
